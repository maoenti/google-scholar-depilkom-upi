```python
import json
import csv

# Buka json
nomor = 1
for count in range(5):
    f = open(f'jsonDepilkom({nomor}).json',)
    nomor += 1

    # JSON jadi dictionaryyyyy
    data = json.load(f)

    # Header CSV
    header = ['Title']
    for i in range(5):
        header.append(f'Author{i+1}')
    header.append('Link')
    header.append('Summary')
    header.append('Cited By')

    # Inisialisasi list data untuk import ke csv
    paper_list = [[] for i in range(len(data)*20)]
    author_list = [[] for j in range(len(data)*20)]
    keys = 0
    counter = 0

    for page in range(len(data)): # Iterasi per page
        for i in range(len(data[page])):
            # PRINT TITLE
            paper_list[keys].append(data[page][i]['title'])

            counter = 0
            # PRINT AUTHOR'S NAME
            try:
                for j in range(len(data[page][i]['publication_info']['authors'])):
                    paper_list[keys].append(data[page][i]['publication_info']['authors'][j]['name'])
                    counter += 1
            except KeyError:
                paper_list[keys].append(None)
                counter += 1
            for ulang in range(counter, 5):
                paper_list[keys].append(None)

            # PRINT LINK
            paper_list[keys].append(data[page][i]['link'])
            
            # PRINT SUMMARY
            paper_list[keys].append(data[page][i]['publication_info']['summary'])
            
            # PRINT TOTAL CITED
            try:
                paper_list[keys].append(data[page][i]['inline_links']['cited_by']['total'])
            except KeyError:
                paper_list[keys].append(0)
            keys = keys + 1
    
    if(count == 0):
        with open('scholarNew.csv', 'w', encoding='UTF8', newline='') as cf:
            writer = csv.writer(cf)

            # write the header
            writer.writerow(header)

            # write multiple rows
            writer.writerows(paper_list)
    elif(count != 0):
        with open('scholarNew.csv', 'a', encoding='UTF8', newline='') as cf:
            writer = csv.writer(cf)

            # write multiple rows
            writer.writerows(paper_list)

    # Closing file
    f.close()
```
